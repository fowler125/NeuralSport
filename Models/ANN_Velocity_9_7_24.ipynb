```python
test = keras.Input(shape = (3,))
```

```python
dense = layers.Dense(64, activation="relu")
x = dense(test)
```

```python
#x = layers.Dense(64, activation="relu")(x)
StrikeZone = layers.Dense(13)(x)
```

```python
model = keras.Model(inputs=test, outputs=StrikeZone, name="mnist_model")
```

```python
#x_train = X_train.reshape(31669, 3).astype("float32")
#x_test = X_test.reshape(6786, 3).astype("float32")



model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.RMSprop(),
    metrics=["accuracy"],
)

history = model.fit(X_train, y_train, epochs=100)

test_scores = model.evaluate(X_test, y_test, verbose=2)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])
```

```python
info = model.fit(X_train, y_train, epochs = 100, validation_data=(X_valid, y_valid))
```

```python
acc = info.history['accuracy']
val_acc = info.history['val_accuracy']

loss = info.history['loss']
val_loss = info.history['val_loss']

plt.figure(figsize=(8, 8))

plt.subplot(2, 1, 1)
plt.plot(acc, label='training data')
plt.plot(val_acc, label='validation data')
plt.ylim([0.3,0.5])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(loss, label='training data')
plt.plot(val_loss, label='validation data')
plt.ylim([1.0,2.5])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend();
```
